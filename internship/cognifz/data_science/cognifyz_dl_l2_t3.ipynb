```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Dataset  (1).csv")
df.columns
```

```python
# Task 1: Extract additional features

df['Restaurant Name Length'] = df['Restaurant Name'].apply(len)
df['Address Length'] = df['Address'].apply(len)
```

```python
# Task 2: Create new features

df['Has Table Booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})
df['Has Online Delivery'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})
```

```python
# Visualize the distributions of the new features
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Restaurant Name Length'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Restaurant Name Length')
plt.xlabel('Length of Restaurant Name')
plt.ylabel('Count')
```

```python
plt.subplot(1, 2, 2)
sns.histplot(df['Address Length'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Address Length')
plt.xlabel('Length of Address')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Visualize the relationship between the new features and existing variables
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='Has Table Booking', y='Aggregate rating', data=df)
plt.title('Average Rating by Table Booking Availability')

plt.subplot(1, 2, 2)
sns.boxplot(x='Has Online Delivery', y='Aggregate rating', data=df)
plt.title('Average Rating by Online Delivery Availability')

plt.tight_layout()
plt.show()
```
