```python
numerical_columns = ['Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']
basic_stats = df[numerical_columns].describe()

print('Basic Statistical Measures:')
print(basic_stats)
```

```python
country_distribution = df['Country Code'].value_counts()
print('Distribution of Country Code:')
country_distribution
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.countplot(y='Country Code', data=df, order=df['Country Code'].value_counts().index)
plt.title('Distribution of Country Code')
plt.xlabel('Count')
plt.ylabel('Country Code')
plt.show()
```

```python
city_distribution = df['City'].value_counts()
print('Distribution of City:')
city_distribution

```

```python
plt.figure(figsize=(12, 6))
top_cities = df['City'].value_counts().head(10)
sns.countplot(y='City', data=df, order=top_cities.index)
plt.title('Top 20 Cities by Number of Restaurants')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

```python
cuisine_distribution = df['Cuisines'].value_counts()
print('Distribution of Cuisines:')
print(cuisine_distribution)
```

```python
plt.figure(figsize=(12, 6))
top_cuisines = df['Cuisines'].value_counts().head(10)
sns.countplot(y='Cuisines', data=df, order=top_cuisines.index)
plt.title('Top 20 Cuisines by Number of Restaurants')
plt.xlabel('Count')
plt.ylabel('Cuisines')
plt.show()
```
